 *Funo para capturar entrada do usurio* 

```python
# Captura uma entrada do usurio
def entradaUsuario(nOpcoes):
  while True:
    try:
      opcao = int(input('\nDigite uma opo: '))
      while opcao > nOpcoes or opcao < 1:
        opcao = int(input(f'Opo invlida! Digite um nmero de 1 a {nOpcoes}: '))
      break
    except:
      print('\n' + f'Opo invlida! A resposta deve ser um nmero inteiro de 1 a {nOpcoes}...'.center(50))
      continue

  return opcao
```

: *Funo para imprimir o cabealho de acordo com o menu que o usurio est* 

```python
# Imprime na tela o cabealho de acordo com qual menu o usurio est
def cabecalho(qualMenu):
  print(colored('_' * 50, 'yellow'))
  print('\n' + colored(qualMenu.center(50), 'blue'))
  print(colored('_' * 50, 'yellow'))
```

 *Funes para imprimir os menus de incio, o menu principal, o menu de compras de produtos e o menu de solicitao de servios* 

```python
# Imprime na tela o menu inicial
def menuInicio():
  while True:
    cabecalho('LOJO DA INFORMTICA')

    print('\n' + 'Seja bem-vindo ao Lojo da Informtica!'.center(50))
    print('O que o senhor(a) deseja fazer?\n'.center(50))

    print(colored('[A] - Acessar o programa'.center(48), 'yellow'))
    print(colored('[F] - Finalizar o programa'.center(49), 'yellow'))
  
    try:
      opcao = input('\nDigite uma opo: ')
      while opcao.upper() != 'A' and opcao.upper() != 'F':
        opcao = input(f'Opo invlida! Digite A ou F: ')
    except:
      print('\n' + f'Opo invlida! A resposta deve ser A ou F...'.center(50))
      continue

    if opcao.upper() == 'A':
      print('\n' + colored('Indo para o menu principal...'.center(50), 'green'))
      menuPrincipal()
      break
    elif opcao.upper() == 'F':
      print('\n' + colored('Obrigado e volte sempre!'.center(50), 'green'))
      break

# Imprime o menu principal
def menuPrincipal():
  cabecalho('MENU PRINCIPAL')

  print('\n' + 'O que o senhor(a) deseja fazer?'.center(50) + '\n')
  print(f'Carrinho: R$ {precoFinal}'.center(50) + '\n')

  print(colored('[1] - Ir ao menu de produtos'.center(50), 'yellow'))
  print(colored('[2] - Ir ao menu de servios'.center(50), 'yellow'))
  print(colored('[3] - Checar carrinho'.center(44), 'yellow'))
  print(colored('[4] - Finalizar compra'.center(43), 'yellow'))
  print(colored('[5] - Sair'.center(32), 'yellow'))

  while True:
    opcao = entradaUsuario(5)
      
    if opcao == 1:
      print('\n' + colored('Maravilha! Indo ao menu de produtos...'.center(50), 'green'))
      menuProdutos()
      break
    elif opcao == 2:
      print('\n' + colored('Legal! Indo ao menu de servios...'.center(50), 'green'))
      menuServicos()
      break
    elif opcao == 3:
      print('\n' + colored('Perfeito! Checando o carrinho...'.center(50), 'green'))
      checarCarrinho()
      menuPrincipal()
      break
    elif opcao == 4:
      print('\n' + colored('Tudo bem! Indo ao menu de pagamentos...'.center(50), 'green'))
      menuPagamentos()
      break
    elif opcao == 5:
      print('\n' + colored('Obrigado e volte sempre! '.center(50), 'green'))
      break

# Imprime o menu de compra de produtos
def menuProdutos():
  cabecalho('MENU DE PRODUTOS')

  print('\n' + 'Aqui est nosso catlogo de produtos'.center(50))
  print('Qual produto o senhor(a) deseja comprar?'.center(50) + '\n')

  produtos = [{'Nome':'Mouse', 'Preco':50},
              {'Nome':'Mouse Pad', 'Preco':20},
              {'Nome':'Teclado', 'Preco':80},
              {'Nome':'Monitor', 'Preco':400},
              {'Nome':'Gabinete', 'Preco':280},
              {'Nome':'Placa Me', 'Preco':450},
              {'Nome':'Placa de Vdeo', 'Preco':1000},
              {'Nome':'Memria RAM 4GB', 'Preco':150},
              {'Nome':'Processador', 'Preco':900},
              {'Nome':'HD', 'Preco':300},
              {'Nome':'Headset', 'Preco':250}]
  
  print('-' * 40)
  print("{:<10} {:<20} {:<15}".format('Opo','Produto','Preo'))
  print('-' * 40)

  for i in range(len(produtos)):
    print(colored("{:<10} {:<20} {:<15}".format(f'[{i+1}]', produtos[i]['Nome'], 'R$ ' + str(produtos[i]['Preco'])), 'yellow'))
  print('-' * 40)
  print(colored("{:<10} {:<20}".format(f'[{len(produtos) + 1}]', 'Sair'), 'yellow'))

  escolha = entradaUsuario(len(produtos) + 1)

  if escolha == (len(produtos) + 1):
    print('\n' + colored('Indo para o menu principal...'.center(50), 'green'))
    menuPrincipal()
  else:
    qtd = escolhaQtd()
    atualizaValor(qtd * produtos[escolha - 1]['Preco'])
    atualizaCarrinho(produtos[escolha - 1], qtd)
    menuPrincipal()

# Imprime o menu de soliticao de servios
def menuServicos():
  cabecalho('MENU DE SERVIOS')

  print('\n' + 'Aqui est nossas opes de servios'.center(50))
  print('Qual servio o(a) senhor(a) deseja?'.center(50) + '\n')

  servicos = [{'Nome':'Formatao', 'Preco':70},
              {'Nome':'Backup', 'Preco':50},
              {'Nome':'Limpeza', 'Preco':80},
              {'Nome':'Reviso', 'Preco':40},
              {'Nome':'Trocar fonte', 'Preco':40},
              {'Nome':'Trocar HD', 'Preco':65}]
  
  print('-' * 40)
  print("{:<10} {:<20} {:<15}".format('Opo','Produto','Preo'))
  print('-' * 40)

  for i in range(len(servicos)):
    print(colored("{:<10} {:<20} {:<15}".format(f'[{i+1}]', servicos[i]['Nome'], 'R$ ' + str(servicos[i]['Preco'])), 'yellow'))
  print('-' * 40)
  print(colored("{:<10} {:<20}".format(f'[{len(servicos) + 1}]', 'Sair'), 'yellow'))

  escolha = entradaUsuario(len(servicos) + 1)

  if escolha == (len(servicos) + 1):
    print('\n' + colored('Indo para o menu principal...'.center(50), 'green'))
    menuPrincipal()
  else:
    qtd = escolhaQtd()
    atualizaValor(qtd * servicos[escolha - 1]['Preco'])
    atualizaCarrinho(servicos[escolha - 1], qtd)
    menuPrincipal()
```

 *Funo para menu de pagamentos* 

```python
# Imprime o menu de escolha da forma de pagamento
def menuPagamentos():
  global precoFinal
  global carrinho

  if precoFinal == 0:
    print()
    print(colored('Seu carrinho est vazio!'.center(50), 'red'))
    menuPrincipal()
    return

  cabecalho('MENU DE PAGAMENTOS')
  print('\n' + 'Qual ser a forma de pagamento do senhor(a)?'.center(50) + '\n')
  print(f'Carrinho: R$ {precoFinal}'.center(50) + '\n')

  print(colored('[1] - Dinheiro'.center(35), 'yellow'))
  print(colored('[2] - Carto de crdito'.center(46), 'yellow'))
  print(colored('[3] - Carto de dbito'.center(43), 'yellow'))
  print(colored('[4] - Pix'.center(32), 'yellow')) 
  print(colored('[5] - Desistir da compra'.center(46), 'yellow'))

  opcao = entradaUsuario(5)

  if opcao == 1:
    cedulasTroco()
    checarCarrinho()
    precoFinal = 0
    carrinho.clear()
    menuInicio()
  elif opcao == 2 or opcao == 3:
    print('\n' + 'Pagamento efetuado com sucesso!'.center(50))
    print('\n' + colored('Obrigado e volte sempre! '.center(50), 'green'))
    checarCarrinho()
    precoFinal = 0
    carrinho.clear()
    menuInicio()
  elif opcao == 4:
    print('\n' + 'Chave pix: 702.858.624-30'.center(50))
    print('\n' + colored('Obrigado e volte sempre! '.center(50), 'green'))
    checarCarrinho()
    precoFinal = 0
    carrinho.clear()
    menuInicio()
  elif opcao == 5:
    print('\n' + colored('Tem certeza que deseja cancelar o pedido? [S/N]'.center(50), 'red'))
    checarCarrinho()

    while True:
      try:
        resp = input('\nDigite sua resposta: ')
        while resp.upper() != 'S' and resp.upper() != 'N':
          resp = int(input(f'Opo invlida! Digite "S" ou "N": '))
        break
      except:
        print('\n' + f'Opo invlida! A resposta deve ser "S" ou "N"...'.center(50))
        continue

    if resp.upper() == 'S':
      print('\n' + 'Compra cancelada com sucesso.'.center(50))
      precoFinal = 0
      carrinho.clear()
      menuPrincipal()
    elif resp.upper() == 'N':
      print('\n' + 'Sua compra no foi cancelada.'.center(50))
      menuPrincipal()
```

 *Funo para clculo do troco* 

```python
# Calcula quanto ser o troco do usurio e quantas cdulas ele ir receber
def cedulasTroco():
  global precoFinal

  while True:
    try:
      trocoPraQuanto = int(input('\nVoc quer troco pra quanto? R$ '))
      while trocoPraQuanto < precoFinal:
        trocoPraQuanto = int(input(f'Opo invlida! O valor tem que ser maior ou igual ao valor total: '))
      break
    except:
      print('\n' + f'Opo invlida! A resposta deve ser um nmero inteiro maior ou igual ao valor total...'.center(50))
      continue

  resto = trocoPraQuanto - precoFinal
  qtdCedulas = 0
  print()

  print(f'Seu troco  de R$ {trocoPraQuanto - precoFinal}')
  while resto != 0:
    if resto >= 50:
      qtdCedulas += (resto // 50)
      print('Cdulas de R$ 50,00: ', resto // 50)
      resto = resto % 50
    elif resto >= 20:
      qtdCedulas += (resto // 20)
      print('Cdulas de R$ 20,00: ', resto // 20)
      resto = resto % 20
    elif resto >= 10:
      qtdCedulas += (resto // 10)
      print('Cdulas de R$ 10,00: ', resto // 10)
      resto = resto % 10
    elif resto >= 5:
      qtdCedulas += (resto // 5)
      print('Cdulas de R$ 5,00: ', resto // 5)
      resto = resto % 5
    elif resto >= 2:
      qtdCedulas += (resto // 2)
      print('Cdulas de R$ 2,00: ', resto // 2)
      resto = resto % 2
    elif resto >= 1:
      qtdCedulas += (resto // 1)
      print('Cdulas de R$ 1,00: ', resto // 1)      
      resto = resto % 1
    
  print('Quantidade total de cdulas para troco: ', qtdCedulas)
  print('\n' + colored('Obrigado e volte sempre! '.center(50), 'green'))
```

 *Funes para atualizar e checar carrinho de compras* 

```python
# Atualizar o carrinho
def atualizaCarrinho(produtoOuServico, qtd):
  global carrinho
  carrinho.append(produtoOuServico)
  carrinho[len(carrinho) - 1]['Quantidade'] = qtd

# Mostra ao usurio os itens que ele escolheu e o valor total
def checarCarrinho():
  global carrinho
  global precoFinal
  print()

  if precoFinal == 0:
    print(colored('Seu carrinho est vazio!'.center(50), 'red'))
    return

  print('-' * 40)
  print("{:<15} {:<15} {:<15}".format('Quantidade','Produto','Preo'))
  print('-' * 40)

  for i in range(len(carrinho)):
    print(colored("{:<15} {:<15} {:<15}".format(carrinho[i]['Quantidade'], carrinho[i]['Nome'], 'R$ ' + str(carrinho[i]['Preco'])), 'yellow'))
  print('-' * 40)
  print("{:<15} {:<20}".format('Valor total do carrinho: R$', str(precoFinal)))
```

 *Funo para atualizar o valor da conta do usurio* 

```python
# Atualiza o valor da conta do usurio
def atualizaValor(valor):
  global precoFinal
  precoFinal += valor
```

 *Funo para o usurio escolher a quantidade de produto ou servio que ele quer* 

```python
# Define a quantidade do produto que o usurio quer
def escolhaQtd():
  while True:
    try:
      qtd = int(input('\nDigite quantas unidades voc quer: '))
      while qtd < 1:
        qtd = int(input(f'Opo invlida! Digite um nmero maior que zero: '))
      break
    except:
      print('\n' + f'Opo invlida! A quantidade deve ser um nmero inteiro maior que zero...'.center(50))
      continue

  return qtd
```
